```python
#Cargar desde un csv sin indice
Delivery=pd.read_csv('Delivery.csv')
```

```python
Delivery
```

```python
#1
Vars_Indep= Delivery[['number_of_available_trolleys','serving_customer', 'maximum_number_of_vehicles_in_the_traffic_jam']]
Var_Dep= Delivery['number_of_operators']
```

```python
#se define modelo como la función de regresión lineal
from sklearn.linear_model import LinearRegression
model= LinearRegression()
```

```python
#Verificamos la función relacionada al modelo 
type(model)
```

```python
#ajustamos el modelo con las variables antes declaradas
model.fit(X=Vars_Indep, y=Var_Dep)
```

```python
#Verificamos los coeficientes obtenidos para el modelo ajustado
model.__dict__
```

```python
#Evaluamos la eficiencia del modelo obtenido por medio del coeficiente R determinado
model.score(Vars_Indep,Var_Dep)
```
